```python
spot_rate_1 = 0.06
spot_rate_2 = 0.07
spot_rate_3 = 0.08
```

```python
annual_rate = 0.07
facial_value = 100
```

```python
present_value = (
    (facial_value * annual_rate) / (1 + spot_rate_1)
    + (facial_value * annual_rate) / (1 + spot_rate_2) ** 2
    + (facial_value * annual_rate) / (1 + spot_rate_3) ** 3
    + facial_value / (1 + spot_rate_3) ** 3
)
print(f"El valor presente del bono es: {present_value:.2f}")
```

```python
from scipy.optimize import brentq


def bond_price(r):
    return (
        (facial_value * annual_rate) / (1 + r)
        + (facial_value * annual_rate) / (1 + r) ** 2
        + (facial_value * annual_rate + facial_value) / (1 + r) ** 3
    ) - present_value


irr = brentq(bond_price, 0, 1)
print(f"La TIR del bono es: {irr:.4%}")
```
