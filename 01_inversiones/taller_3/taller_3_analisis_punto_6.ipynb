# Punto 6 — Tasa Libre de Riesgo (^COLCAP)


```python
import yfinance as yf
import numpy as np

# Descarga datos históricos del COLCAP
colcap = yf.download("ICOLCAP.CL", start="2021-01-01", auto_adjust=True, progress=False)

# Retornos diarios logarítmicos
daily_returns = np.log(colcap["Close"] / colcap["Close"].shift(1)).dropna()

# Retorno promedio anualizado (252 días de trading)
avg_daily_return = daily_returns.mean().item()
avg_annual_return = avg_daily_return * 252

print(
    f"Retorno promedio diario:    {avg_daily_return:.6f}  ({avg_daily_return * 100:.4f}%)"
)
print(
    f"Retorno promedio anualizado: {avg_annual_return:.6f}  ({avg_annual_return * 100:.4f}%)"
)
print(f"\nTasa de mercado estimada (COLCAP): {avg_annual_return * 100:.2f}% anual")
```
